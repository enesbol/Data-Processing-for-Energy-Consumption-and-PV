```python
junehourly = pd.read_csv("JuneHourly.csv")
junehourly.drop(['Unnamed: 0'],axis=1,inplace=True)


def JuneSpace(junehourly,house):
    deneme = junehourly[junehourly['Username']==house]
    forindex = deneme[deneme['Datatype'] == 'kWhReturnedLow']
    listforretlow = deneme[deneme['Datatype'] == 'kWhReturnedLow'].Waarde.values
    listforretnormal = deneme[deneme['Datatype'] == 'kWhReturnedNormal'].Waarde.values
    listforusednormal = deneme[deneme['Datatype'] == 'kWhUsedNormal'].Waarde.values
    listforusedlow = deneme[deneme['Datatype'] == 'kWhUsedLow'].Waarde.values
    listforpv = deneme[deneme['Datatype'] == 'PV_kWh'].Waarde.values
    
    calcfor1stspace = listforpv - listforretnormal-listforretlow + listforusednormal+listforusedlow

    dffor1stspace = pd.DataFrame(calcfor1stspace)
    dffor1stspace.index = pd.to_datetime(forindex['Tijdstip'])
    dffor1stspace['PV'] = calcfor1stspace*10
    dffor1stspace.columns=['Energy_Consumption','PV']
    return dffor1stspace

```

```python
def SeptSpace(septhourly,house):
    deneme = septhourly[septhourly['Username']==house]
    forindex = deneme[deneme['Datatype'] == 'kWhReturnedLow']
    listforretlow = deneme[deneme['Datatype'] == 'kWhReturnedLow'].Waarde.values
    listforretnormal = deneme[deneme['Datatype'] == 'kWhReturnedNormal'].Waarde.values
    listforusednormal = deneme[deneme['Datatype'] == 'kWhUsedNormal'].Waarde.values
    listforusedlow = deneme[deneme['Datatype'] == 'kWhUsedLow'].Waarde.values
    listforpv = deneme[deneme['Datatype'] == 'PV_kWh'].Waarde.values
    
    calcfor1stspace = listforpv - listforretnormal-listforretlow + listforusednormal+listforusedlow

    dffor1stspace = pd.DataFrame(calcfor1stspace)
    dffor1stspace.index = pd.to_datetime(forindex['Tijdstip'])
    dffor1stspace['PV'] = calcfor1stspace
    dffor1stspace.columns=['Energy_Consumption','PV']
    return dffor1stspace
```

```python
#df0 = pd.read_csv("threepart/first_part.csv") #188 days 18048 - 15min

#df.drop(['Unnamed: 0'],axis=1,inplace=True)

def CalculateThree(df,house):
  
    Returned_Low_Normal = df[(df['Username']==house)]
    Returned_Low_Normal['Tijdstip'] = pd.to_datetime(Returned_Low_Normal['Tijdstip'])
    days = (Returned_Low_Normal.Tijdstip.iloc[-1] - Returned_Low_Normal.Tijdstip.iloc[0]).days
    norm = days*96 
    if "Property" in Returned_Low_Normal.columns:
        Returned_Normal_Df = Returned_Low_Normal[(Returned_Low_Normal['Property']== 'kWhReturnedNormal')]
        Returned_Low_Df = Returned_Low_Normal[(Returned_Low_Normal['Property']== 'kWhReturnedLow')]
    elif "Datatype" in Returned_Low_Normal.columns:
        Returned_Normal_Df = Returned_Low_Normal[(Returned_Low_Normal['Datatype']== 'kWhReturnedNormal')]
        Returned_Low_Df = Returned_Low_Normal[(Returned_Low_Normal['Datatype']== 'kWhReturnedLow')]
        
    Returned_Normal_Df['Returned_Normal_Values_Diff'] = Returned_Normal_Df['Waarde'].diff()
    Returned_Normal_Df['Returned_Normal_Time_Diff'] = Returned_Normal_Df['Tijdstip'].diff()
    
    Returned_Low_Df['Returned_Low_Values_Diff'] = Returned_Low_Df['Waarde'].diff()
    Returned_Low_Df['Returned_Low_Time_Diff'] = Returned_Low_Df['Tijdstip'].diff()


    Returned_Low_Df.Returned_Low_Values_Diff.iloc[0] = Returned_Low_Df.Returned_Low_Values_Diff.iloc[1]
    Returned_Low_Df.Returned_Low_Time_Diff.iloc[0] = Returned_Low_Df.Returned_Low_Time_Diff.iloc[1]


    Returned_Normal_Df.Returned_Normal_Values_Diff.iloc[0] = Returned_Normal_Df.Returned_Normal_Values_Diff.iloc[1]
    Returned_Normal_Df.Returned_Normal_Time_Diff.iloc[0] = Returned_Normal_Df.Returned_Normal_Time_Diff.iloc[1]

    List_For_Returned_Low = []
    for i in range(0,Returned_Low_Df.shape[0]):
        a = (Returned_Low_Df.Returned_Low_Time_Diff.iloc[i].seconds / 3600)
        List_For_Returned_Low.append(a)

    Returned_Low_Df['Hourly_Time_Diff'] =List_For_Returned_Low
    Returned_Low_Df['Calculation'] = Returned_Low_Df['Returned_Low_Values_Diff'] / Returned_Low_Df['Hourly_Time_Diff']


    List_For_Returned_Normal = []
    for i in range(0,Returned_Normal_Df.shape[0]):
        b = (Returned_Normal_Df.Returned_Normal_Time_Diff.iloc[i].seconds / 3600)
        List_For_Returned_Normal.append(b)

    Returned_Normal_Df['Hourly_Time_Diff'] =List_For_Returned_Normal
    Returned_Normal_Df['Calculation'] = Returned_Normal_Df['Returned_Normal_Values_Diff'] / Returned_Normal_Df['Hourly_Time_Diff']

    quant = Returned_Normal_Df['Hourly_Time_Diff'].quantile(0.15)
    Returned_Normal_Df.drop(Returned_Normal_Df[Returned_Normal_Df['Hourly_Time_Diff'] < quant].index,inplace=True)

    quant = Returned_Low_Df['Hourly_Time_Diff'].quantile(0.15)
    Returned_Low_Df.drop(Returned_Low_Df[Returned_Low_Df['Hourly_Time_Diff'] < quant].index, inplace = True)
 ### ### ### ### ### ### ### ### ### ### ### ### ### ### ### ### ### ### ### ### ### ### ### ### ### ### ### ### ### ###
    if (Returned_Normal_Df.shape[0] > norm)  or (Returned_Normal_Df.shape[0] < norm ):
        Returned_Normal_Df_Normalizing = Returned_Normal_Df[['Calculation']]
        Returned_Normal_Df_Normalizing.index = pd.to_datetime(Returned_Normal_Df['Tijdstip'])
        Returned_Normal_Df_Normalizing = Returned_Normal_Df_Normalizing.asfreq('15min',method="ffill")

    elif (Returned_Normal_Df.shape[0] == norm):
        Returned_Normal_Df_Normalizing = Returned_Normal_Df


    if (Returned_Low_Df.shape[0] > norm ) or (Returned_Low_Df.shape[0] < norm ):
        Returned_Low_Df_Normalizing = Returned_Low_Df[['Calculation']]
        Returned_Low_Df_Normalizing.index = pd.to_datetime(Returned_Low_Df['Tijdstip'])
        Returned_Low_Df_Normalizing = Returned_Low_Df_Normalizing.asfreq('15min' , method='ffill')

    elif (Returned_Low_Df.shape[0] == norm):
        Returned_Low_Df_Normalizing = Returned_Low_Df
        
    Returned_Normal_list = Returned_Normal_Df_Normalizing.Calculation.tolist()
    Returned_Low_list = Returned_Low_Df_Normalizing.Calculation.tolist()

    if len(Returned_Low_list) == len(Returned_Normal_list):
        pass
    else:
        if len(Returned_Normal_list) < norm :
            Returned_Normal_list.append(Returned_Normal_list[-1])


        if len(Returned_Low_list) < norm :
            Returned_Low_list.append(Returned_Low_list[-1])

    Calculations_For_Returned = [x + y for x, y in zip(Returned_Normal_list, Returned_Low_list)]

    if (len(Calculations_For_Returned) < norm ):
        Calculations_For_Returned.append(Calculations_For_Returned[-1])

    Used_Low_Normal = df[(df['Username']==house)]
    Used_Low_Normal['Tijdstip'] = pd.to_datetime(Used_Low_Normal['Tijdstip'])
    if "Property" in Used_Low_Normal.columns:
        Used_Normal_Df = Used_Low_Normal[(Used_Low_Normal['Property']== 'kWhUsedNormal')]
        Used_Low_Df = Used_Low_Normal[(Used_Low_Normal['Property']== 'kWhUsedLow')]

    elif "Datatype" in Used_Low_Normal.columns:
        Used_Normal_Df = Used_Low_Normal[(Used_Low_Normal['Datatype']== 'kWhUsedNormal')]
        Used_Low_Df = Used_Low_Normal[(Used_Low_Normal['Datatype']== 'kWhUsedLow')]
        
    Used_Normal_Df['Used_Normal_Values_Diff'] = Used_Normal_Df['Waarde'].diff()
    Used_Normal_Df['Used_Normal_Time_Diff'] = Used_Normal_Df['Tijdstip'].diff()

    Used_Low_Df['Used_Low_Values_Diff'] = Used_Low_Df['Waarde'].diff()
    Used_Low_Df['Used_Low_Time_Diff'] = Used_Low_Df['Tijdstip'].diff()
    
    
    
    Used_Normal_Df.Used_Normal_Values_Diff.iloc[0] = Used_Normal_Df.Used_Normal_Values_Diff.iloc[1]
    Used_Normal_Df.Used_Normal_Time_Diff.iloc[0] = Used_Normal_Df.Used_Normal_Time_Diff.iloc[1]

    Used_Low_Df.Used_Low_Values_Diff.iloc[0] = Used_Low_Df.Used_Low_Values_Diff.iloc[1]
    Used_Low_Df.Used_Low_Time_Diff.iloc[0] = Used_Low_Df.Used_Low_Time_Diff.iloc[1]
    

    List_For_Used_Normal = []
    for i in range(0,Used_Normal_Df.shape[0]):
        a = (Used_Normal_Df.Used_Normal_Time_Diff.iloc[i].seconds / 3600)
        List_For_Used_Normal.append(a)

    Used_Normal_Df['Hourly_Time_Diff'] = List_For_Used_Normal
    Used_Normal_Df['Calculation'] = Used_Normal_Df['Used_Normal_Values_Diff'] / Used_Normal_Df['Hourly_Time_Diff']
    
    List_For_Used_Low = []
    for i in range(0,Used_Low_Df.shape[0]):
        a = (Used_Low_Df.Used_Low_Time_Diff.iloc[i].seconds / 3600)
        List_For_Used_Low.append(a)

    Used_Low_Df['Hourly_Time_Diff'] =List_For_Used_Low
    Used_Low_Df['Calculation'] = Used_Low_Df['Used_Low_Values_Diff'] / Used_Low_Df['Hourly_Time_Diff']
    
    Used_Low_Df = Used_Low_Df[~Used_Low_Df.Tijdstip.duplicated(keep='first')] 
    Used_Normal_Df = Used_Normal_Df[~Used_Normal_Df.Tijdstip.duplicated(keep='first')] 
    
    
    if (Used_Low_Df.shape[0] > norm) or (Used_Low_Df.shape[0] < norm ):
        Used_Low_Df_Normalizing = Used_Low_Df[['Calculation']]
        Used_Low_Df_Normalizing.index = pd.to_datetime(Used_Low_Df['Tijdstip'])
        Used_Low_Df_Normalizing = Used_Low_Df_Normalizing.asfreq('15min' , method='ffill')
    elif (Used_Low_Df.shape[0] == norm):
        Used_Low_Df_Normalizing = Used_Low_Df
        
    if (Used_Normal_Df.shape[0] > norm ) or (Used_Normal_Df.shape[0] < norm ):
        Used_Normal_Df_Normalizing = Used_Normal_Df[['Calculation']]
        Used_Normal_Df_Normalizing.index = pd.to_datetime(Used_Normal_Df['Tijdstip'])
        Used_Normal_Df_Normalizing = Used_Normal_Df_Normalizing.asfreq('15min' , method='ffill')
    elif (Used_Normal_Df.shape[0] == norm):
        Used_Normal_Df_Normalizing = Used_Normal_Df
        
    Used_Normal_list = Used_Normal_Df_Normalizing.Calculation.tolist()
    Used_Low_list = Used_Low_Df_Normalizing.Calculation.tolist()
    

    if len(Used_Low_list) == len(Used_Normal_list):
        pass

    else:

        if len(Used_Normal_list) < norm :
            Used_Normal_list.append(Used_Normal_list[-1])
        if len(Used_Low_list) < norm :
            Used_Low_list.append(Used_Low_list[-1])  
    
    Calculations_For_Used = [x + y for x, y in zip(Used_Normal_list, Used_Low_list)]
    
    if (len(Calculations_For_Used) < norm ):
        Calculations_For_Used.append(Calculations_For_Used[-1])
        
    if "Property" in df.columns:
        For_Pv = df[(df['Username'] == house) & (df['Property'] == 'PV_kWh')]
    elif "Datatype" in df.columns:
        For_Pv = df2 [(df['Username'] ==house) & (df['Datatype'] == 'PV_kWh')]
        
    For_Pv['Tijdstip'] = pd.to_datetime(For_Pv['Tijdstip'])

    arr = np.diff(For_Pv.Waarde.values)
    arr = np.insert(arr, 0 ,arr[0])
    
    For_Pv['PV_kWh_Values_Diff'] = arr
    For_Pv['PV_kWh_Time_Diff'] = For_Pv['Tijdstip'].diff()
    
    For_Pv.PV_kWh_Values_Diff.iloc[0] = For_Pv.PV_kWh_Values_Diff.iloc[1]
    For_Pv.PV_kWh_Time_Diff.iloc[0] = For_Pv.PV_kWh_Time_Diff.iloc[1]


    For_Pv_Values = []
    for i in range(0, For_Pv.shape[0]):
        a = (For_Pv.PV_kWh_Time_Diff.iloc[i].seconds / 3600)
        For_Pv_Values.append(a)

    For_Pv['PV_kWh_Time_Diff'] = For_Pv_Values
    For_Pv['Calculation'] = For_Pv['PV_kWh_Values_Diff'] / For_Pv['PV_kWh_Time_Diff']

    quant = For_Pv['PV_kWh_Time_Diff'].quantile(0.15)
    For_Pv.drop(For_Pv[For_Pv['PV_kWh_Time_Diff'] < quant].index, inplace = True)



    if (For_Pv.shape[0] > norm ) or (For_Pv.shape[0] < norm ):

        For_Pv_Normalizing = For_Pv[['Calculation']]
        For_Pv_Normalizing.index = pd.to_datetime(For_Pv['Tijdstip'])
        For_Pv_Normalizing = For_Pv_Normalizing[~For_Pv_Normalizing.index.duplicated(keep='first')]
        For_Pv_Normalizing = For_Pv_Normalizing.asfreq('15min', method="ffill")

    if (For_Pv.shape[0] == norm):
        For_Pv_Normalizing = For_Pv	

    Calculations_For_Pv = For_Pv_Normalizing['Calculation'].values

    if (len(Calculations_For_Pv) < norm ):
        Calculations_For_Pv.append(Calculations_For_Pv[-1])	

    
    if len(Calculations_For_Returned) < len(Calculations_For_Pv):
        Calculations_For_Returned.append(Calculations_For_Returned[-1])

    if len(Calculations_For_Used) < len(Calculations_For_Pv):
        Calculations_For_Used.append(Calculations_For_Used[-1])

    if len(Calculations_For_Pv) < len(Calculations_For_Pv):
        Calculations_For_Pv.append(Calculations_For_Pv[-1])
    
    #Final=[]
    #for i in range(0,len(Calculations_For_Returned)):
        #Final.append(Calculations_For_Returned[i] - Calculations_For_Used[i])
    
    Final =  Calculations_For_Pv - Calculations_For_Returned + Calculations_For_Used

    dffinal = pd.DataFrame(Final)
    dffinal.index = For_Pv_Normalizing.index
    dffinal.columns=['Energy_Consumption']
    dffinal["PV"] = Calculations_For_Pv

    return dffinal 
    
 
```
